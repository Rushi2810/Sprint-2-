# Overall Attrition Rate
### Explanation: Calculates the percentage of employees who left the organization

```python
query_overall_attrition = """
SELECT 
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data;
"""
# Execute the query and fetch the result
overall_attrition = pd.read_sql(query_overall_attrition, con=engine)
print("Overall Attrition Rate:")
print(overall_attrition)
```

### Calculates the attrition rate department wise

```python
query_department_attrition = """
SELECT 
    Department, 
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY Department
ORDER BY Attrition_Rate DESC;
"""
# Execute the query and fetch the result
department_attrition = pd.read_sql(query_department_attrition, con=engine)
print("\nDepartment-Wise Attrition Rate:")
print(department_attrition)
```

### Calculates the attrition rate for male and female employees

```python
query_gender_attrition = """
SELECT 
    Gender, 
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY Gender;
"""
gender_attrition = pd.read_sql(query_gender_attrition, con=engine)
print("Attrition by Gender:")
print(gender_attrition)
```

### Calculates the attrition rate for employees based on marital status (Single, Married, Divorced)

```python
query_marital_status_attrition = """
SELECT 
    MaritalStatus, 
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY MaritalStatus;
"""
marital_status_attrition = pd.read_sql(query_marital_status_attrition, con=engine)
print("\nAttrition by Marital Status:")
print(marital_status_attrition)
```

###  Categorizes employees into age groups and calculates attrition rate for each group

```python
query_age_attrition = """
SELECT 
    CASE 
        WHEN Age BETWEEN 20 AND 30 THEN '20-30'
        WHEN Age BETWEEN 31 AND 40 THEN '31-40'
        WHEN Age BETWEEN 41 AND 50 THEN '41-50'
        WHEN Age > 50 THEN '50+'
    END AS Age_Group,
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY Age_Group;
"""
age_attrition = pd.read_sql(query_age_attrition, con=engine)
print("\nAttrition by Age Groups:")
print(age_attrition)
```

## Satisfaction Insights

### Average Satisfaction Scores for Attrition Groups

```python
query_satisfaction_attrition = """
SELECT 
    Attrition,
    AVG(JobSatisfaction) AS Avg_JobSatisfaction,
    AVG(EnvironmentSatisfaction) AS Avg_EnvironmentSatisfaction,
    AVG(RelationshipSatisfaction) AS Avg_RelationshipSatisfaction
FROM employee_data
GROUP BY Attrition;
"""
satisfaction_attrition = pd.read_sql(query_satisfaction_attrition, con=engine)
print("Average Satisfaction Scores by Attrition Status:")
print(satisfaction_attrition)
```

### Visualization: Satisfaction Comparison

```python
# Prepare data for bar chart
labels = ['Job Satisfaction', 'Environment Satisfaction', 'Relationship Satisfaction']
attrition_yes = satisfaction_attrition[satisfaction_attrition['Attrition'] == 'Yes'][['Avg_JobSatisfaction', 'Avg_EnvironmentSatisfaction', 'Avg_RelationshipSatisfaction']].values.flatten()
attrition_no = satisfaction_attrition[satisfaction_attrition['Attrition'] == 'No'][['Avg_JobSatisfaction', 'Avg_EnvironmentSatisfaction', 'Avg_RelationshipSatisfaction']].values.flatten()

x = range(len(labels))  # Positions for the bars

# Plot satisfaction comparison
plt.figure(figsize=(10, 6))
plt.bar(x, attrition_yes, width=0.4, label='Attrition = Yes', color='red', align='center')
plt.bar([p + 0.4 for p in x], attrition_no, width=0.4, label='Attrition = No', color='green', align='center')
plt.xticks([p + 0.2 for p in x], labels)
plt.ylabel('Average Satisfaction Score')
plt.title('Average Satisfaction Scores by Attrition Status')
plt.legend()
plt.show()
```



### Key insight:
- Employees who left (`Attrition = Yes`) consistently reported lower satisfaction levels across all categories:
  - **Job Satisfaction**: 2.47 vs. 2.78.
  - **Environment Satisfaction**: 2.46 vs. 2.77.
  - **Relationship Satisfaction**: 2.60 vs. 2.73.
- This highlights the critical role of satisfaction in retention, with **job and environment satisfaction** showing the largest gaps.


#  Work-Life Balance

### Work-Life Balance by Attrition

```python
query_wlb_attrition = """
SELECT 
    WorkLifeBalance,
    COUNT(*) AS Total_Employees,
    SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Attrition_Count,
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY WorkLifeBalance
ORDER BY WorkLifeBalance;
"""
wlb_attrition = pd.read_sql(query_wlb_attrition, con=engine)
print("Work-Life Balance Impact on Attrition:")
print(wlb_attrition)
```

### Work-Life Balance by Department

```python
query_wlb_department = """
SELECT 
    Department,
    WorkLifeBalance,
    COUNT(*) AS Total_Employees,
    SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Attrition_Count,
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY Department, WorkLifeBalance
ORDER BY Department, WorkLifeBalance;
"""
wlb_department = pd.read_sql(query_wlb_department, con=engine)
print("\nWork-Life Balance Across Departments:")
print(wlb_department)
```

### Visualization: Attrition by Work-Life Balance

```python
plt.figure(figsize=(10, 6))
plt.bar(wlb_attrition['WorkLifeBalance'], wlb_attrition['Attrition_Rate'], color='blue')
plt.xlabel('Work-Life Balance Rating')
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Work-Life Balance Rating')
plt.xticks([1, 2, 3, 4], labels=['Bad', 'Good', 'Better', 'Best'])
plt.show()
```

### Key Insights:

- Employees with a **Work-Life Balance rating of 1 ("Bad")** have the **highest attrition rate** (over 30%), showing a significant negative impact of poor work-life balance.
- Attrition rates decrease as work-life balance improves:
  - **Rating 2 ("Good")**: Moderate attrition rate (~15%).
  - **Rating 3 ("Better")**: Lower attrition rate (~10%).
  - **Rating 4 ("Best")**: Slightly higher attrition (~12%), potentially due to other factors despite a good work-life balance.

### Visualization: Work-Life Balance by Department

```python
# Prepare the pivot table for heatmap-like representation
wlb_department_pivot = wlb_department.pivot(index='Department', columns='WorkLifeBalance', values='Attrition_Rate')

# Plot heatmap for department-wise work-life balance
plt.figure(figsize=(12, 8))
plt.imshow(wlb_department_pivot, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Attrition Rate (%)')
plt.xticks(range(len(wlb_department_pivot.columns)), labels=['Bad', 'Good', 'Better', 'Best'], rotation=45)
plt.yticks(range(len(wlb_department_pivot.index)), labels=wlb_department_pivot.index)
plt.title('Attrition Rate by Work-Life Balance and Department')
plt.xlabel('Work-Life Balance Rating')
plt.ylabel('Department')
plt.show()
```

### Key insight
- **Sales** shows the highest attrition rate for poor work-life balance ratings (1 and 2), emphasizing the need for work-life balance improvements in this department.
- **Human Resources** also exhibits moderate attrition rates for lower work-life balance ratings, suggesting potential issues with workload or role expectations.
- **Research & Development** demonstrates relatively stable and low attrition rates across all work-life balance ratings, indicating better management of work-life balance.

### Work-Life Balance by Job Role

```python
query_wlb_jobrole = """
SELECT 
    JobRole,
    WorkLifeBalance,
    COUNT(*) AS Total_Employees,
    SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Attrition_Count,
    (SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS Attrition_Rate
FROM employee_data
GROUP BY JobRole, WorkLifeBalance
ORDER BY JobRole, WorkLifeBalance;
"""
wlb_jobrole = pd.read_sql(query_wlb_jobrole, con=engine)
print("Work-Life Balance Impact on Attrition by Job Role:")
print(wlb_jobrole)


```

```python
# Visualization: Grouped Bar Chart
# Prepare data for plotting
job_roles = wlb_jobrole['JobRole'].unique()
wlb_levels = [1, 2, 3, 4]  # WorkLifeBalance levels
wlb_colors = ['red', 'orange', 'blue', 'green']  # Colors for levels

# Plotting grouped bar chart
plt.figure(figsize=(14, 8))
width = 0.2  # Bar width
for i, wlb in enumerate(wlb_levels):
    wlb_data = wlb_jobrole[wlb_jobrole['WorkLifeBalance'] == wlb]
    plt.bar(
        [x + i * width for x in range(len(job_roles))],
        wlb_data['Attrition_Rate'],
        width=width,
        label=f'Work-Life Balance {wlb}',
        color=wlb_colors[i]
    )

plt.xticks([x + 1.5 * width for x in range(len(job_roles))], job_roles, rotation=45)
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Work-Life Balance and Job Role')
plt.legend(title='Work-Life Balance Rating')
plt.show()
```

### Key insight:
- **Laboratory Technicians**:
  - Employees with a **rating of 1 ("Bad")** have an extremely high attrition rate (over 60%), making this role highly sensitive to poor work-life balance.
- **Sales Representatives**:
  - High attrition across all ratings, with a notable spike at **rating 4 ("Best")**, indicating other factors influencing attrition.
- **Research Scientists and Healthcare Representatives**:
  - Generally lower attrition rates, even for poor work-life balance ratings, suggesting better adaptability or job satisfaction.
- **Executives and Managers**:
  - Show moderate attrition rates, but they handle work-life balance better compared to other roles.
